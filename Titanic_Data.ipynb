```python
def age_fill(data):
    Age = data[0]
    Pclass = data[1]
    if pd.isnull(Age) == True:
        if Pclass == 1:
            return 37
        if Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
            
```

```python
titanic['Age'] = titanic[['Age','Pclass']].apply(age_fill,axis=1)
```

```python
test['Age'] = test[['Age','Pclass']].apply(age_fill,axis=1)
```

```python
titanic.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```
